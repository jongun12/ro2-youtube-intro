```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0, 6, 0.01)
```

```python
y = np.sin(np.pi * t)
```

```python
plt.figure(figsize=(12,6))
plt.plot(t,y)
plt.grid()
plt.show()
```

```python
def draw_sin(t,A,f,b):
    y = A * np.sin(np.pi * f * t) + b

    plt.figure(figsize=(12,6))
    plt.plot(t,y)
    plt.grid()
    plt.show()
```

```python
t = np.arange(0, 4, 0.01)
draw_sin(t, 2, 10, 2)
```

```python
class DrawSin:
    def __init__(self, amp, freq, bias, end_time):
        self.amp = amp
        self.freq = freq
        self.bias = bias
        self.end_time = end_time

    def calc_sin(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.sin(2*np.pi*self.freq*self.t) + self.bias
    
    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize=(12,6))
        plt.plot(self.t,y)
        plt.grid()
        plt.show()
```

```python
ds = DrawSin(1, 1, 0, 3)
ds.__dict__
```

```python
ds.draw_sin()
```

```python
class DrawSinusoidal(DrawSin):
    def calc_cos(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.cos(2*np.pi*self.freq*self.t) + self.bias
    
    def draw_cos(self):
        y = self.calc_cos()
        plt.figure(figsize=(12,6))
        plt.plot(self.t,y)
        plt.grid()
        plt.show()
```

```python
dc = DrawSinusoidal(1, 1, 0, 3)
dc.__dict__
```

```python
dc.__dir__()
```

```python
dc.draw_cos()
```

```python
dc.draw_sin()
```
